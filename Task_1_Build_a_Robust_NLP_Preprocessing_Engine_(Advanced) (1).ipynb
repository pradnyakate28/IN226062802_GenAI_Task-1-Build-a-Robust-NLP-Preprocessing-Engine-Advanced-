Q1. What is the difference between "Love" and "love" in NLP?                             
In NLP, "Love" and "love" are treated as different words because of case sensitivity. Usually, we convert all text to lowercase so both become "love" and are treated the same.                                                      




Q2. What happens if stopwords are not removed?     
If stopwords are not removed, the data will contain many unnecessary words like "is", "the", "and". This increases data size and may reduce model accuracy because these words do not add much meaning.





Q3. Provide two real-world scenarios where removing stopwords can be harmful.                   
a)In sentiment analysis, removing words like "not" can change the meaning (e.g., "not good" becomes "good").                                                    
b)In chatbots or question answering systems, removing stopwords may affect understanding of the sentence.



Q4. Explain the difference between stemming and lemmatization.                                        
Stemming reduces words to a base form by cutting them (e.g., "running" → "run" or "runn").
Lemmatization converts words into their correct base form using meaning (e.g., "running" → "run").

```python
#Remove very short tokens
text = "I am not happy with this"
print(f"Original Text : {text}")

def preprocess_text(text):
    words = text.split()  # tokenize

    filtered_words = []
    for word in words:
        if len(word) > 2 or word in ['no', 'not']:
            filtered_words.append(word)

    return filtered_words, " ".join(filtered_words)

tokens, processed = preprocess_text(text)

print(f"Tokens : {tokens}")
print(f"Processed Text : {processed}")
```

```python
# Final preprocessing function
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\S+|\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)
    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()
    tokens = [w for w in tokens if len(w) > 2 or w in ['no', 'not']]

    clean_sentence = " ".join(tokens)
    return tokens, clean_sentence

# Test sentences
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

# Output format
for sentence in test_sentences:
    tokens, clean_sentence = preprocess_text(sentence)

    print("Original Text :", sentence)
    print("Cleaned Tokens :", tokens)
    print("Cleaned Sentence :", clean_sentence)
    print("-" * 50)
```

```python
import numpy as np

for sentence in test_sentences:
    tokens, clean_sentence = preprocess_text(sentence)

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))

    # average token length
    if total_tokens > 0:
        avg_length = sum(len(word) for word in tokens) / total_tokens
    else:
        avg_length = 0

    print("Sentence :", sentence)
    print("Total Tokens :", total_tokens)
    print("Unique Tokens :", unique_tokens)
    print("Average Token Length :", round(avg_length, 2))
    print("-" * 50)
```

```python
from collections import Counter

# Combine all tokens
all_tokens = []

for sentence in test_sentences:
    tokens, _ = preprocess_text(sentence)
    all_tokens.extend(tokens)

# Create frequency counter
counter = Counter(all_tokens)

# Top 10 most frequent words
print("Top 10 Most Frequent Words:")
print(counter.most_common(10))

print("-" * 50)

# Top 5 least frequent words
print("Top 5 Least Frequent Words:")
print(counter.most_common()[-5:])
```

```python
##TASK 6 :  Build Full Pipeline
```

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean_sentence = preprocess_text(text)

        all_tokens.extend(tokens)              # combine all tokens
        clean_sentences.append(clean_sentence) # store cleaned sentence

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
result = full_pipeline(test_sentences)

print("All Tokens:")
print(result["tokens"])

print("\nClean Sentences:")
print(result["clean_sentences"])
```

```python
##TASK 7 : Error Handling
```

```python
def preprocess_text(text):
    try:
        # Handle empty or invalid input
        if not text or not isinstance(text, str):
            return [], ""

        # Convert to lowercase
        text = text.lower()

        # Remove URLs and emails
        text = re.sub(r'http\S+|www\S+|\S+@\S+', '', text)

        # Remove numbers
        text = re.sub(r'\d+', '', text)

        # Remove special characters (including emojis)
        text = re.sub(r'[^\w\s]', '', text)

        # Handle repeated characters (e.g., loooove → loove)
        text = re.sub(r'(.)\1{2,}', r'\1\1', text)

        # Remove extra spaces
        text = re.sub(r'\s+', ' ', text).strip()

        # Tokenization
        tokens = text.split()

        # Remove short tokens (≤2) except 'no', 'not'
        tokens = [w for w in tokens if len(w) > 2 or w in ['no', 'not']]

        # Final cleaned sentence
        clean_sentence = " ".join(tokens)

        return tokens, clean_sentence

    except Exception:
        return [], ""

#Edge Case Testing
test_cases = [
    "",                 # Empty string
    "😍😍😍",           # Only emojis
    "1234567890"        # Only numbers
]

for case in test_cases:
    tokens, clean_sentence = preprocess_text(case)

    print("Input:", case)
    print("Tokens:", tokens)
    print("Clean Sentence:", clean_sentence)
    print("-" * 40)
```
